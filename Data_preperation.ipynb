##Choosing the subset of data we need


Since airports have pretty accurate weather data, we are going to choose Seattle Tacome Airport station and JFK International Airport as the stations that we explore. in order to be consistent we'll only work with these two stations and these stations will represent the city.

```python
df_seatac = df_seattle.loc[df_seattle["NAME"] == "SEATTLE TACOMA AIRPORT, WA US"]
```

```python
df_jfk = df_ny.loc[df_ny["NAME"] == "JFK INTERNATIONAL AIRPORT, NY US"]
```

```python
df_jfk.head()
```

```python
df_seatac.head()
```

Next We'll clear any duplicates from the data

```python
df_jfk = df_jfk.drop_duplicates(subset="DATE")
df_seatac = df_seatac.drop_duplicates(subset="DATE")
```

And see if we have a row for each date for the past 3 years (we should have 1461 rows : 365 * 4 + 1(leap year)).

```python
df_jfk.shape
```

```python
df_seatac.shape
```

We can see that jfk data has a row for every date in the past 4 years. but seatac data does not.

First we will join the data and then we fill in the missing data for Seatac


##Joining the Data

```python
df = df_jfk[['DATE', 'PRCP']].merge(df_seatac[['DATE', 'PRCP']], on='DATE', how='left')
```

We do a left join so we can keep all the information from df_jfk and only the non null PRCP from the seatac dataset

```python
df
```

We'll create a tidy version of the dataset


```python
df = pd.melt(df, id_vars = 'DATE', var_name = 'CITY', value_name = 'PRCP')
df
```

then we'll rename the perciption to reflect the name of the cities

```python
df.loc[df['CITY']=='PRCP_x', 'CITY'] = 'NYC'
df.loc[df['CITY']=='PRCP_y', 'CITY'] = 'SEA'
```

```python
df.head()
```

```python
df.tail()
```

We can rename the columns to follow the lower case rule in order for them to be more readable

```python
mapping_dict = {'DATE':'date', 'CITY':'city', 'PRCP':'precipitation'}
```

```python
df = df.rename(columns=mapping_dict)
```

##Filling in the missing data

```python
df["precip_interp_2"] = df["precipitation"]
df_sea = df.loc[df["city"] == "SEA"]
df_sea["day_of_year"] = df_sea["date"].dt.dayofyear
nan_idx = pd.isna(df_sea["precipitation"])
df_mean_daily = df_sea.groupby("day_of_year")["precipitation"].mean()
```

```python
nan_idx
```

```python
df_sea.loc[nan_idx, "precip_interp_2"] = df_mean_daily.loc[df_sea.loc[nan_idx, "day_of_year"].values].values
```

```python
df.loc[df_sea.index] = df_sea
```

```python
df["precipitation"] = df["precip_interp_2"]
df = df.drop(columns=["precip_interp_2"])
```

```python
df
```

in the code above we get the average precipitation from the previous years and replace that average with the missing data

Add new variables month and year to use in analysis later.

```python
df['month'] = pd.DatetimeIndex(df['date']).month
df["year"] = df["date"].dt.year
```
